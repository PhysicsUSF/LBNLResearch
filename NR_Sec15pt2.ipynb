```python
'''
Date: 1/10/15

This is a simplest exercise in chi2-fitting there are four methods of doing it below.  

All four methods agree on the best-fit (m, b) and their uncertainties (two from NR, one based on grid search, 
and the last one using scipy_fit().

(It can be used as an exercise in Computational Physics.)


Using NR eq 15.2.4 and eqs 15.2.15 - 15.2.22 to compute the best-fit slope and intercept for a straight line and the
variances and covariance.

Ultimate goal: to understand whether the p-values in NR's table on 697 are linked to chi2 or reduced chi2
'''

## constructing data
steps = 100
xstart = 1.
xend = 2.
xd = linspace(xstart, xend, steps)
m = 3.
b = 1.5
sig = 1.
noise = sig*random.randn(xd.shape[0])
#noise = sqrt(xd)*random.randn(xd.shape[0])
varz = noise**2
yd = m*xd + b + noise
#errorbar(xd, yd, noise, 'k.')
errorbar(xd, yd, noise, fmt='r.', markersize=10, label=u'Observations')

print 'variance', var(yd - (xd*m + b))

## Figuring out best-fit (m, b) and the covariance matrix using NR section 15.2 example.

## auxiliary quantities 
## Method 1 - See NR eq 15.2.4, eq 15.2.15 - 15.2.21
S = sum(1/varz)
Sx = sum(xd/varz)
Sy = sum(yd/varz)
t = (xd - Sx/S)/noise
Stt = sum(t**2)

## Method 2 - NR eq 15.2.6.  According to NR, this method has larger numerical error.
Sxx = sum(xd**2/varz)
Sxy = sum(xd*yd/varz)
Delta = S*Sxx - Sx**2

## best-fit (m, b) and covariance matrix.

## Method 1 
m_fit = sum(t*yd/noise)/Stt
b_fit = (Sy - Sx*m)/S

var_m = 1/Stt
var_b = (1 + Sx**2/(S*Stt))/S
cov_mb = -Sx/(S*Stt)

print 'True m, b:', m, b
print 'm_fit, b_fit:', m_fit, b_fit
print 'var_m, var_b, cov_mb:', var_m, var_b, cov_mb
print 'sig_m, sig_b:', sqrt(var_m), sqrt(var_b)


## Method 2
b_fit2 = (Sxx*Sy - Sx*Sxy)/Delta
m_fit2 = (S*Sxy - Sx*Sy)/Delta

var_m2 = S/Delta
var_b2 = Sxx/Delta
cov_mb2 = -Sx/Delta

print 'm_fit2, b_fit2:', m_fit2, b_fit2
print 'var_m2, var_b2, cov_mb2:', var_m2, var_b2, cov_mb2


## chi2_min
dof = steps - 2.
chi2_min = sum((yd - m_fit*xd - b_fit)**2/varz)
red_chi2_min = chi2_min/dof
print 'chi2_min, red_chi2_min', chi2_min, red_chi2_min

if red_chi2_min > 1.5:
    print '***************************************'
    print 'WARNING: Large red_chi2_min:', red_chi2_min
    print '***************************************'
## plotting best-fit based on Method 1

yfit = xd*m_fit + b_fit
plot(xd, yfit, 'k--', linewidth = 4)

show()

```

```python
'''

This has conclusively shown the Table on NR p. 697 is for chi2 and NOT reduced chi2!!!  1/11/15
This also shows the chi2 way and the probability way of plotting the contours are indeed the same.

'''


##### chi2 grid.  

from copy import deepcopy
from photom_vs_spectral3D import compute_sigma_level

m_steps = 100
del_m = .3
b_steps = 100
del_b = .5

m_arr = linspace(m_fit - del_m, m_fit + del_m, m_steps)
b_arr = linspace(b_fit - del_b, b_fit + del_b, b_steps)

#print 'm_arr', m_arr
#print 'b_arr', b_arr


### Figuring out minimum chi2.
chi2 = np.zeros((len(m_arr), len(b_arr)))
                
max_m_1sig, min_m_1sig = m_fit, m_fit
max_b_1sig, min_b_1sig = b_fit, b_fit
for i, m_try in enumerate(m_arr):
    for j, b_try in enumerate(b_arr):
        chi2[i, j] = np.sum((yd - m_try*xd - b_try)**2/varz)

dof = steps - 2.
red_chi2 = chi2/dof        
chi2_min = chi2.min()
mindex = where(chi2 == chi2_min)
print 'from the grid'
m_fit3 = m_arr[mindex[0]][0]
b_fit3 = b_arr[mindex[1]][0]
print 'best-fit m and b:', m_fit3, b_fit3 

red_chi2_min_grid = chi2_min/dof
print 'red_chi2_min from the grid:', red_chi2_min_grid

### Find one sigma uncertainty based on NR's Table on p. 697.
max_m_1sig, min_m_1sig, max_m_2sig, min_m_2sig = m_fit3, m_fit3, m_fit3, m_fit3
max_b_1sig, min_b_1sig, max_b_2sig, min_b_2sig = b_fit3, b_fit3, b_fit3, b_fit3


best_line = m_fit3*xd + b_fit3
snake_hi_1sig = deepcopy(best_line)
snake_lo_1sig = deepcopy(best_line)


for i, m_try in enumerate(m_arr):
    for j, b_try in enumerate(b_arr):
        del_chi2 = chi2[i, j] - chi2_min
        if del_chi2 < 4.0:
            max_m_2sig = np.maximum(max_m_2sig, m_try)
            min_m_2sig = np.minimum(min_m_2sig, m_try)
            max_b_2sig = np.maximum(max_b_2sig, b_try)
            min_b_2sig = np.minimum(min_b_2sig, b_try)
            

            if del_chi2 < 1.0:
                max_m_1sig = np.maximum(max_m_1sig, m_try)
                min_m_1sig = np.minimum(min_m_1sig, m_try)
                max_b_1sig = np.maximum(max_b_1sig, b_try)
                min_b_1sig = np.minimum(min_b_1sig, b_try)

        if del_chi2 < 18.4:           
            try_line = m_try*xd + b_try
            snake_hi_1sig = np.maximum(snake_hi_1sig, try_line)  # the result is the higher values from either array is picked.
            snake_lo_1sig = np.minimum(snake_lo_1sig, try_line)
            #print 'm_try, b_try', m_try, b_try
            #print 'snake_hi_1sig - best_line', snake_hi_1sig - best_line
            #print 'snake_lo_1sig - best_line', snake_lo_1sig - best_line
            #raise KeyboardInterrupt #equiv to Ctrl-C.  The only way I know how to stop running in notebook 
                                        #(exit(1) will cause kernel to be shut down and one has to restart every time.


print 'from the grid:'
print 'max_m_1sig, min_m_1sig', max_m_1sig, min_m_1sig
print 'max_b_1sig, min_b_1sig', max_b_1sig, min_b_1sig
print 'max_m_2sig, min_m_2sig', max_m_2sig, min_m_2sig
print 'max_b_2sig, min_b_2sig', max_b_2sig, min_b_2sig


sig_m_hi = max_m_1sig - m_fit3
sig_m_lo = m_fit3 - min_m_1sig
sig_m = (sig_m_hi + sig_m_lo)/2
sig_b_hi = max_b_1sig - b_fit3
sig_b_lo = b_fit3 - min_b_1sig
sig_b = (sig_b_hi + sig_b_lo)/2

two_sig_m_hi = max_m_2sig - m_fit3
two_sig_m_lo = m_fit3 - min_m_2sig
two_sig_m = (two_sig_m_hi + two_sig_m_lo)/2
two_sig_b_hi = max_b_2sig - b_fit3
two_sig_b_lo = b_fit3 - min_b_2sig
two_sig_b = (two_sig_b_hi + two_sig_b_lo)/2



print 'sig_m_hi, sig_m_lo, sig_m: %.5f, %.5f, %.5f' % (sig_m_hi, sig_m_lo, sig_m)
print 'sig_b_hi, sig_b_lo, sig_b: %.4f, %.4f, %.4f' % (sig_b_hi, sig_b_lo, sig_b)

print 'Checking if the 2-sigma uncertainties (from chi2 criterion) are indeed twice the 1-sigma uncertainties:'
print 'two_sig_m_hi, two_sig_m_lo, two_sig_m: %.5f, %.5f, %.5f' % (two_sig_m_hi, two_sig_m_lo, two_sig_m)
print 'two_sig_b_hi, two_sig_b_lo, two_sig_b: %.4f, %.4f, %.4f' % (two_sig_b_hi, two_sig_b_lo, two_sig_b)

#### plotting contours

### plotting contours using probability
P = exp(-chi2/2)
print 'Without proper normalization factor, sum(P) is NOT 1:', sum(P)
P = P/sum(P)
CDF = compute_sigma_level(P).T

X, Y = meshgrid(m_arr, b_arr)
contour_levels = [0.0, 0.683, 0.955, 1.0]

## why plot 1-CDF and then set the levels to be 1 - levels??  Becaues of the zero??
contourf(X, Y, 1-CDF, levels=[1-l for l in contour_levels], cmap=mpl.cm.summer, alpha=0.5)
plot([m_arr.min(), m_arr.max()], [max_b_1sig, max_b_1sig], 'k--')
plot([m_arr.min(), m_arr.max()], [min_b_1sig, min_b_1sig], 'k--')
plot([max_m_1sig, max_m_1sig], [b_arr.min(), b_arr.max()], 'b--')
plot([min_m_1sig, min_m_1sig], [b_arr.min(), b_arr.max()], 'b--')

#plt.contour(X, Y, CHISQ-chisq_min, levels=[1.0, 4.0], colors=['r', 'g'])
        
## mark minimum
scatter(m_fit3, b_fit3, marker='s', facecolors='r')
#show()

### plotting contours using chi2

chi2_contour_levels = [0.0, 2.3, 6.17]
del_chi2 = chi2 - chi2_min
## I don't understand the transpose done for del_chi but it seems to be the right thing to do: 
## recall the CDF is also transposed.
contour(X, Y, del_chi2.T, levels = chi2_contour_levels, linewidth = 3)
plot([m_arr.min(), m_arr.max()], [max_b_1sig, max_b_1sig], 'k--')
plot([m_arr.min(), m_arr.max()], [min_b_1sig, min_b_1sig], 'k--')
plot([max_m_1sig, max_m_1sig], [b_arr.min(), b_arr.max()], 'b--')
plot([min_m_1sig, min_m_1sig], [b_arr.min(), b_arr.max()], 'b--')

#plt.contour(X, Y, CHISQ-chisq_min, levels=[1.0, 4.0], colors=['r', 'g'])
        
## mark minimum
scatter(m_fit3, b_fit3, marker='s', facecolors='r')
show()



#### plot error snakes.
f, ax = subplots()
plot(xd, best_line, 'k--')
errorbar(xd, yd, noise, fmt='r.', markersize=10, label=u'Observations')
ax.fill_between(xd, snake_hi_1sig, snake_lo_1sig, facecolor='black', alpha=0.5)
#ax.fill_between(reg_wave, , , facecolor='black', alpha=0.3)
show()
```

```python
'''
A fourth way of determining the fitting parameters and their uncertainties -- using scipy.curve_fit()

All four methods agree on the best-fit (m, b) and their uncertainties (two from NR, one based on grid, and the last one
using scipy_fit()

'''

from scipy.optimize import curve_fit
def func(x, m, b):
    return m * x + b


popt, pcov = curve_fit(func, xd, yd, p0 = [2.9, 1.3], sigma = noise, absolute_sigma=True)

print 'optimal m and b', popt
print '1-sigma uncertainties on m and b', np.sqrt(np.diag(pcov))
print 'covariance matrix between m and b', pcov
```
